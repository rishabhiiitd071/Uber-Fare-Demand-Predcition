```python
ride_share_df = pd.read_csv("rideshare_kaggle.csv")
ride_share_df.info()
```

```python
ride_share_df.describe()
```

```python
ride_share_df=ride_share_df[ride_share_df['source'] != ride_share_df['destination']]
```

```python
ride_share_df['price']= ride_share_df['price'].apply(pd.to_numeric, errors='coerce')
ride_share_df['price']= ride_share_df['price'].replace({0:np.nan})
ride_share_df['distance']= ride_share_df['distance'].apply(pd.to_numeric, errors='coerce')
ride_share_df['distance']= ride_share_df['distance'].replace({0:np.nan})
ride_share_df['surge_multiplier']= ride_share_df['surge_multiplier'].apply(pd.to_numeric, errors='coerce')
ride_share_df['surge_multiplier']= ride_share_df['surge_multiplier'].replace({0:np.nan})
ride_share_df['latitude']= ride_share_df['latitude'].apply(pd.to_numeric, errors='coerce')
ride_share_df['latitude']= ride_share_df['latitude'].replace({0:np.nan})
ride_share_df['longitude']= ride_share_df['longitude'].apply(pd.to_numeric, errors='coerce')
ride_share_df['longitude']= ride_share_df['longitude'].replace({0:np.nan})
```

```python
ride_share_df.columns[ride_share_df.isnull().any()]
```

```python
ride_share_df['price'] = ride_share_df['price'].fillna(ride_share_df['price'].mean())
```

```python
values=['price', 'distance','surge_multiplier']

linear_Data = ride_share_df[values]
linear_Data
```

```python
#separate the other attributes from the predicting attribute
x = linear_Data.drop("price",axis=1)
#separte the predicting attribute into Y for model training 
y = linear_Data["price"]
```

```python
# importing train_test_split from sklearn
from sklearn.model_selection import train_test_split
# splitting the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel = RandomForestRegressor(n_estimators=100, random_state=101)
```

```python
rfrmodel.fit(x_train,y_train)
rfrmodel_pred= rfrmodel.predict(x_test)
```

```python
print("R2 score using Random Forest Regressor is", r2_score(y_test, rfrmodel_pred))
print('Mean Squared Error of Random Forest Regressor is',mean_squared_error(y_test, rfrmodel_pred))
print('Root Mean Squared Error of Random Forest Regressor is',np.sqrt(mean_squared_error(y_test, rfrmodel_pred)))
```

```python
loc_coord = {'Haymarket Square' : [42.3628, -71.0583],
             'North Station' : [42.3664, -71.0620],
             'Northeastern University' : [42.3398, -71.0892],
             'West End' : [42.4556, -75.0966],
             'South Station' : [42.3519, -71.0552],
             'Fenway' : [42.3519, -71.0552],
             'Theatre District' : [40.7590, -73.9845],
             'Beacon Hill' : [42.3588, -71.0707],
             'Back Bay' : [42.3503, -71.0810],
             'North End' : [42.3647, -71.0542],
             'Financial District' : [40.7077, -74.0083],
             'Boston University' : [42.3505, -71.1054]
}
```

```python
src_latitude = []
src_longitude = []
dest_latitude = []
dest_longitude = []

for i in range(len(ride_share_df)):
    src_latitude.append(loc_coord[ride_share_df['source'][i]][0])
    src_longitude.append(loc_coord[ride_share_df['source'][i]][1])
    dest_latitude.append(loc_coord[ride_share_df['destination'][i]][0])
    dest_longitude.append(loc_coord[ride_share_df['destination'][i]][1])

ride_share_df['pickup_latitude'] = src_latitude
ride_share_df['pickup_longitude'] = src_longitude
ride_share_df['dropoff_latitude'] = dest_latitude
ride_share_df['dropoff_longitude'] = dest_longitude
```

```python
rel_cols = ['datetime', 'hour', 'day', 'month', 'source', 'destination', 'distance', 'price', 'pickup_latitude', 'pickup_longitude', 'dropoff_latitude', 'dropoff_longitude']
```

```python
new_data = ride_share_df[rel_cols]
new_data['datetime'] = new_data['datetime'].astype(str)
```

```python
new_data.head(5)
```

```python
df_train = new_data.sample(frac=0.8,random_state=200)
df_test = new_data.drop(df_train.index, axis=0)
```

```python
df_test1 = df_test.copy(deep = True)
```

```python
df_train, df_validation = train_test_split(df_train, test_size=0.10, random_state=1)
```

```python
validation_target = df_validation['price'].values
```

```python
def clean_dataframe(df):
        
    df = df[(df['dropoff_longitude'] != df['pickup_longitude'])]
    df = df[(df['dropoff_latitude'] != df['pickup_latitude'])]
    
    return df
  
    
def distance(pickup_lat, pickup_long, dropoff_lat, dropoff_long):
    return np.abs(dropoff_lat - pickup_lat) + np.abs(dropoff_long - pickup_long)   
    
def add_features(df):
    plat = df['pickup_latitude']
    dlat = df['dropoff_latitude']
    plon = df['pickup_longitude']
    dlon = df['dropoff_longitude']


    df['latitude_diff'] = (plat - dlat)
    df['longitude_diff'] = (plon - dlon)
    
    df['datetime'] = df['datetime'].str.slice(0, 16)
    df['datetime'] = pd.to_datetime(df['datetime'], utc=True, format='%Y-%m-%d %H:%M')
    
    df['hour'] = df['datetime'].dt.hour
    df['day'] = df['datetime'].dt.day
    df['month'] = df['datetime'].dt.month
    df['year'] = df['datetime'].dt.year
    df['weekday'] = df['datetime'].dt.weekday
    df['quarter_of_year'] = df['datetime'].dt.quarter
    
    lgr = (-73.8733, 40.7746)
    jfk = (-73.7900, 40.6437)
    ewr = (-74.1843, 40.6924)

    df['trip_distance_km'] = distance(df.pickup_latitude, df.pickup_longitude, df.dropoff_latitude, df.dropoff_longitude)   
    df['pickup_distance_jfk'] = distance(df['pickup_latitude'], df['pickup_longitude'], jfk[1], jfk[0])
    df['dropoff_distance_jfk'] = distance(df['dropoff_latitude'], df['dropoff_longitude'], jfk[1], jfk[0])
    df['pickup_distance_ewr'] = distance(df['pickup_latitude'], df['pickup_longitude'], ewr[1], ewr[0])
    df['dropoff_distance_ewr'] = distance(df['dropoff_latitude'], df['dropoff_longitude'], ewr[1], ewr[0])
    df['pickup_distance_laguardia'] = distance(df['pickup_latitude'], df['pickup_longitude'], lgr[1], lgr[0])
    df['dropoff_distance_laguardia'] = distance(df['dropoff_latitude'], df['dropoff_longitude'], lgr[1], lgr[0]) 
    
    return df


df_train = clean_dataframe(df_train)
df_validation = clean_dataframe(df_validation)
df_train = add_features(df_train)
df_validation = add_features(df_validation)
df_test = add_features(df_test)


print(df_train.corrwith(df_train['price']))


paratest = df_test[['pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','latitude_diff','longitude_diff','trip_distance_km','pickup_distance_ewr','dropoff_distance_ewr','pickup_distance_laguardia','dropoff_distance_laguardia']]
para = df_train[['pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','latitude_diff','longitude_diff','trip_distance_km','pickup_distance_ewr','dropoff_distance_ewr','pickup_distance_laguardia','dropoff_distance_laguardia']]
paraval = df_validation[['pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','latitude_diff','longitude_diff','trip_distance_km','pickup_distance_ewr','dropoff_distance_ewr','pickup_distance_laguardia','dropoff_distance_laguardia']]

target = df_train.price
validation_target = df_validation.price
test_price = df_test.price

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_train = scaler.fit_transform(para)
df_validation = scaler.transform(paraval)
df_test = scaler.transform(paratest)
```

```python
n_col = para.shape[1]
print(n_col)

from keras.layers import BatchNormalization
from keras.layers.advanced_activations import LeakyReLU

import keras.backend as K

def rmse (y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true), axis=-1))


def model() :   #Created a Model using Keras
    model = Sequential()
    model.add(Dropout(0.2,input_shape=(n_col,)))
    model.add(BatchNormalization())
    model.add(Dense(512,activation='relu'))#512 neurons in input layer
    model.add(Dropout(0.2))
    model.add(BatchNormalization())
    model.add(Dense(256,activation='relu')) #256 neurons in hidden layer
    model.add(BatchNormalization())
    model.add(Dense(128,activation='relu'))  # 128 neurons in hidden layer
    model.add(BatchNormalization())
    model.add(Dense(64,activation='relu'))   # 64 neurons in hidden layer
    model.add(BatchNormalization())
    model.add(Dense(32,activation='relu'))   # 32 neurons in hidden layer
    model.add(BatchNormalization())
    model.add(Dense(16,activation='relu')) # 16 neurons in hidden layer
    model.add(BatchNormalization())
    model.add(Dense(8,activation='relu')) # 8 neurons in hidden layer
    model.add(BatchNormalization())
    model.add(Dense(1)) # 1 neuron in output layer
    
    #nadam = optimizers.Nadam(lr=0.002, beta_1=0.9, beta_2=0.999, epsilon=None, schedule_decay=0.004)
    #adadelta =optimizers.Adadelta(lr=1.0, rho=0.95, epsilon=None, decay=0.0)
    #adgrad = optimizers.Adagrad(lr=0.001, epsilon=None, decay=0.0)
    #rms = optimizers.RMSprop(lr=0.001, rho=0.9, epsilon=None, decay=0.0)
    
    #sgd = optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True,clipnorm=0.5)
    adam = tf.keras.optimizers.Adam(learning_rate=0.001)
    model.compile(optimizer=adam,loss='mean_squared_error', metrics=[rmse])
    return model


model = model()
estimator = model.fit(x=para,y=target, batch_size=1024, epochs = 3, 
                    verbose=1, validation_data=(paraval,validation_target), 
                    shuffle=True)
score = model.evaluate(paratest, test_price, batch_size = 1024)
```
